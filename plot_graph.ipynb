```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
# Import an undirected network
path = 'data/networks/as1998_500.txt'
G = nx.read_edgelist(path, delimiter='\t', nodetype=int)
```

```python
pr = nx.pagerank(G)
```

```python
colors = cm.rainbow(range(len(pr)))
np.random.shuffle(colors)
```

```python
plt.figure(figsize=(10,10))
nx.draw_networkx(G, pos=nx.spring_layout(G, seed = 10, k=1.7, iterations=50), node_size = np.array(list(pr.values()))*10000, linewidth=1, with_labels=False, alpha=0.95, node_color = 'midnightblue', edge_color='dimgray')
```
